# Коэффициент связанного состояния

Асимптотика связанного состояния: $\psi_{\text{bound}} \approx C x^{z/\kappa}\exp(-\kappa x)$

```python
# асимптотика связанного состояния
def psi_bound_asymp(x, C, z, kappa):
    return C*x**(z/kappa)*np.exp(-kappa * x)
```

# Сшивка асимптотики связанного состояния и ВКБ

Обозначения:

$z$ - заряд остова

$\kappa=\sqrt{2I_P}$

Модуль импульса под барьером:  $|p|=\sqrt{\kappa^2-\dfrac{2z}{x^2} - 2Ex}$

```python
# зависимость импульса от координаты
def p_abs(x, z, kappa, E):
    return np.sqrt(np.abs(2*z/np.sqrt(x**2) + 2*E*x -kappa**2))
```

Уравнение на точки поворота: $\kappa^2-\dfrac{2z}{x^2} - 2Ex = 0$

Первое приближение: $~~~x_{in}=\dfrac{z}{I_p} = \dfrac{2z}{\kappa^2}$;
$~~~x_{out}=\dfrac{Ip}{E} = \dfrac{\kappa^2}{2E}$

```python
# точки поворота
def get_x(z, kappa, E):
    x_in = 2*z/kappa**2
    x_out = kappa**2 / (2*E)
    x_in = sp.optimize.root_scalar(lambda x: 2*z/np.sqrt(x**2) + 2*E*x -kappa**2, x0 = x_in)["root"]
    x_out = sp.optimize.root_scalar(lambda x: 2*z/np.sqrt(x**2) + 2*E*x -kappa**2, x0 = x_out)["root"]
    return x_in, x_out
```

Квазиклассическая функция под барьером: $\psi_{sc} = \dfrac{A}{\sqrt{|p|}}\exp\left( -\int\limits_{x_0}^{x} |p| dx \right)$

```python
# квазиклассическая в.ф. под барьером
def psi_sc_under(x, A, z, kappa, E):
    x_in, x_out = get_x(z, kappa, E)
    integral = sp.integrate.quad(lambda x: p_abs(x, z, kappa, E), x_in, x)[0]
    return A/np.sqrt(p_abs(x, z, kappa, E)) * np.exp(-integral)
psi_sc_under = np.vectorize(psi_sc_under)
```

Применимость квазиклассики около левой точки поворота: $\left|\dfrac{d}{dx}\dfrac{1}{p}\right|\ll 1$

$$
\left| \dfrac{z}{x^2\left( \dfrac{2 z}{x} - \kappa^2\right)^{3/2}} \right| \ll 1
$$

```python
def vkb_applty(x, z, kappa):
    return z / ( x**2 * (np.abs(2*z/x - kappa**2))**(3/2) )
```

## Рассмотрим Ar18

```python
x_in, x_out = get_x(z, kappa, E)
print(f"z/Ip = {z/Ip}")
print(f"z/kappa = {z/kappa}")
print(f"aB = {z/kappa**2}")
print(f"x_in = {x_in}")
print(f"x_out = {x_out}")
```

```python
x = np.linspace(1.05*x_in, 0.9*x_out, 100)

x_for_fit = [0.15, 0.6]
print(f"vkb_applty = {vkb_applty(x_for_fit[0], z, kappa)}")
x_fit = np.linspace(x_for_fit[0], x_for_fit[1], 100)
vkb_fit = sp.optimize.curve_fit(lambda x, A: psi_sc_under(x, A, z, kappa, E), x_fit, psi_bound_asymp(x_fit, C_fit, z, kappa))
A_fit = np.abs(vkb_fit[0][0])
print(f"vkb_fit = {vkb_fit}")
print(f"A_fit = {A_fit}")


fig, ax = plt.subplots(1, 1, figsize=(6,4))
ax.plot(x, psi_bound_asymp(x, C_fit, z, kappa), color = "blue", label="psi0_asymp")
ax.plot(x, psi_sc_under(x, A_fit, z, kappa, E), color = "red", label="psi_sc_under")
ax.axvline(x_for_fit[0], color='black', linestyle='-.')
ax.axvline(x_for_fit[1], color='black', linestyle='-.')
# ax.set(xlim=[0.1, 0.3], ylim=[1e-3, 1])
ax.set(yscale = 'log')
ax.grid(); ax.legend();
```
